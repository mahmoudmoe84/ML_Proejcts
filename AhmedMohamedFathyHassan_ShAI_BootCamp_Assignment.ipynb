3. **Data Cleaning**: Handle missing data by suitable method with explain why you use it.


```python
df_drop = df[['Id', 'EmployeeName', 'JobTitle', 'BasePay', 'OvertimePay', 'OtherPay'
, 'TotalPay', 'TotalPayBenefits', 'Year', 'Agency']]
```

```python
df_drop.info()
```

```python
df_drop.isna().sum()
```

**There are many ways to handle missing values:-**

```python
# A-Dropping missing values if the missing values is small and not impacting on the dat
df_drop.dropna(inplace=True)
```

```python
df_drop.isna().sum()
```

```python
df_drop.info()
```

```python
 # B-Imputation: Mean,Median, (for numerical data) - Mode (for categorical data) - constant imputaion(value)
df_imp = df[['Id', 'EmployeeName', 'JobTitle', 'BasePay', 'OvertimePay', 'OtherPay'
, 'TotalPay', 'TotalPayBenefits', 'Year', 'Agency']]
```

```python
df_imp.isna().sum()
```

```python
# Mean imputation
df_imp['BasePay'].fillna(df_imp['BasePay'].mean(), inplace=True)

# Median imputation
df_imp['OvertimePay'].fillna(df_imp['OvertimePay'].median(), inplace=True)

# Constant imputation
value = 5000
df_imp['OtherPay'].fillna(value, inplace=True)
```

```python
df_imp.isna().sum() # there is no missing data
```

4- **Basic Data Visualization:** Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
import matplotlib.pyplot as plt

# Create histogram for salaries
plt.figure(figsize=(8, 6))
plt.hist(df['TotalPay'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Create pie chart for departments by using Pandas
plt.figure(figsize=(20, 20))
df['JobTitle'].value_counts().head(5).plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Proportion of Employees in Different Departments')
plt.ylabel('')
plt.show()
```

```python
# Pie Chart for Department Proportions by using Matplotlib
department_counts = df['JobTitle'].value_counts().head(5)
labels = department_counts.index
sizes = department_counts.values

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue', 'gold'])
plt.title('Proportion of Employees in Different Jobtitle')
plt.show()
```

5- **Grouped Analysis:** Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.



```python
df
```

```python
# Group the data by 'Year' and calculate summary statistics
summary_stats = df.groupby('Year')['TotalPay'].agg(['mean', 'median', 'min', 'max', 'std', 'count'])

# Compare average salaries across different groups
average_salaries = summary_stats.sort_values(by='mean', ascending=False)

print(average_salaries)
```

```python
#visualizing above data using histogram
import seaborn as sns
sns.barplot(x = 'Year', y = 'mean', data = average_salaries, color= 'skyblue')
plt.title('Mean of Salary for Years')
plt.show()
```

```python
#visualizing above data using histogram
import seaborn as sns
sns.barplot(x = 'Year', y = 'std', data = average_salaries, color= 'salmon')
plt.title('STD of Salary for Years')
plt.show()
```

6- **Simple Correlation Analysis:** Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.

```python
# Calculate correlation between 'Salary' and 'Years_Experience'
correlation = df['TotalPay'].corr(df['Year'])

print("Correlation between Salary and Years:", correlation)

# Plot scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['Year'], df['TotalPay'] ,color='blue')
plt.title('Salary vs Years')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
